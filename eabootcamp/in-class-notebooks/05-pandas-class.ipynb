```python
import os
import sqlite3

import earthpy as et
import pandas as pd
```

```python
# Downlaod sqlite file
fire_url = (
    "https://www.fs.usda.gov/rds/archive/products/"
    "RDS-2013-0009.5/RDS-2013-0009.5_SQLITE.zip"
)

et.data.get_data(url=fire_url)
```

```python
# Open sqlite file
os.chdir(os.path.join(et.io.HOME, 'earth-analytics', 'data'))

fire_path = os.path.join(
    'earthpy-downloads',
    'RDS-2013-0009.5_SQLITE',
    'Data',
    'FPA_FOD_20210617.sqlite')

if os.path.exists(fire_path):
    print("There is a file at {}".format(fire_path))
else:
    raise ValueError('No file at {}'.format(fire_path))
```

```python
raw_fire_df = pd.read_sql_query(
    'SELECT * FROM fires',
    'sqlite:///' + fire_path)
raw_fire_df
```

```python
# Get CA fires
raw_ca_fire_df = raw_fire_df[raw_fire_df.STATE=='CA']
raw_ca_fire_df.head()
```

```python
# get fires larger than 100 acres
ca_gt100_fire_df = raw_ca_fire_df[raw_ca_fire_df.FIRE_SIZE>=100]
```

```python
ca_gt100_fire_df1 = raw_fire_df[
    (raw_fire_df.STATE=='CA') & (raw_fire_df.FIRE_SIZE>=100)
]
all(ca_gt100_fire_df1 == ca_gt100_fire_df)
ca_gt100_fire_df
```

```python
fire_df = ca_gt100_fire_df[
    ['FOD_ID','FIRE_YEAR','NWCG_GENERAL_CAUSE', 'FIRE_SIZE']
]
fire_df
```

```python
fire_df = fire_df.rename(columns={
    'FOD_ID': 'id',
    'FIRE_YEAR': 'year',
    'NWCG_GENERAL_CAUSE': 'cause',
    'FIRE_SIZE': 'size'
})
fire_df
```
